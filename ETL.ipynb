<b>Primera modificacion de los CSV: 

Creacion y modificacion de la columna ID<b>

```python
# Creo la columna ID concatenando la info de show_id mas la letra correspondiente a su df (a(Amazon), d(Disney), h(hulu), n(Netflix)). 
df_a['id'] = "a" + df_a['show_id']
df_d['id'] = "d" + df_d['show_id']
df_h['id'] = "h" + df_h['show_id']
df_n['id'] = "n" + df_n['show_id']

# Verifico que se haya creado y aplicado la columna.
df_a.head()
```

<b>Reordeno la columna ID para que quede primera<b>

```python
# Verifico el orden de las columnas.
print('a',df_a.columns)# Printear un .columns del dataframe me permite ver las columnas, copio esas columnas para en el proximo paso cambiarlas de lugar(Lo realizo para no escribir 1 por 1 las columnas)
print('d',df_d.columns)# Tambien me sirve para ver como esta escrita cada columna en el DF si tienen alguna separacion etc.
print('h',df_h.columns)
print('n',df_n.columns)
```

```python
# Reordeno las columnas del DataFrame utilizando la nueva lista de columnas.
df_a = df_a.reindex(columns=['id', 'show_id', 'type', 'title', 'director', 'cast', 'country',
                     'date_added', 'release_year', 'rating', 'duration', 'listed_in',
                     'description'])
df_d = df_d.reindex(columns=['id', 'show_id', 'type', 'title', 'director', 'cast', 'country',
                     'date_added', 'release_year', 'rating', 'duration', 'listed_in',
                     'description'])
df_h = df_h.reindex(columns=['id', 'show_id', 'type', 'title', 'director', 'cast', 'country',
                     'date_added', 'release_year', 'rating', 'duration', 'listed_in',
                     'description'])
df_n = df_n.reindex(columns=['id', 'show_id', 'type', 'title', 'director', 'cast', 'country',
                     'date_added', 'release_year', 'rating', 'duration', 'listed_in',
                     'description'])
```

```python
# Drope la columna show_id ya que fue remplazada por id.
df_a.drop(columns= 'show_id', inplace = True)
df_d.drop(columns= 'show_id', inplace = True)
df_h.drop(columns= 'show_id', inplace = True)
df_n.drop(columns= 'show_id', inplace = True)
```

```python
# Verifico que se realizaron las modificaciones 
print('a',df_a.columns)
print('d',df_d.columns)
print('h',df_h.columns)
print('n',df_n.columns)
```

<b>Segunda modificacion del Dataframe: 

Modificacion de la columna rating<b>

```python
# Verifico cuántos valores nulos hay en la columna "rating" con el .isnull, realizamos un .sum para que los sume.
df_plataforma['rating'].isnull().sum()
```

```python
# Relleno los nulos de la columna con la funcion .fillna con una G
df_plataforma['rating'] = df_plataforma['rating'].fillna('G')
```

<b>Realizo modificaciones en la columna date_added<b>

```python
# Vistazo rapido de los dataset
df_plataforma.info()
```

```python
# Tuve que verificar en la columna date_added la misma se encontraba con una separacion al principio lo cual no permitia ejecutar la linea de comando que le sigue a esta, to_datatime.
df_plataforma['date_added'] = df_plataforma['date_added'].str.strip()
```

```python
# Creo las nuevas columnas "duration_int" y "duration_type" a partir de la de "duration" con un .split a partir del espacio que hay entre ellos en los registros.
df_plataforma[['duration_int', 'duration_type']] = df_plataforma['duration'].str.split(' ', n=1, expand=True)
```

```python
# Utilizo una funcion lambda dentro del applymap para que recorra todo el df y pase a minusculas todos los string
df_plataforma = df_plataforma.applymap(lambda x: x.lower() if type(x) == str else x)
# applymap es una funcion de pandas que se utiliza para aplicar una función elemento por elemento en todo el DataFrame.(En este caso lambda)

# Verifico que se haya aplicado correctamente
df_plataforma.head(1)
```

<b>Union de los CSV RATING

Creacion de la columna SCORE<B>

```python
# Leo los archivos en la carpeta ratings para trabajar con los mismos.
df1= pd.read_csv('ratings/1.csv')
df2= pd.read_csv('ratings/2.csv')
df3= pd.read_csv('ratings/3.csv')
df4= pd.read_csv('ratings/4.csv')
df5= pd.read_csv('ratings/5.csv')
df6= pd.read_csv('ratings/6.csv')
df7= pd.read_csv('ratings/7.csv')
df8= pd.read_csv('ratings/8.csv')
```

```python
# Un vistazo de la cantidad de registros que contiene cada uno para darme una idea del volumnes que voy a trabajar.
print("df1",df1.shape)
print("df2",df2.shape)
print("df3",df3.shape)
print("df4",df4.shape)
print("df5",df5.shape)
print("df6",df6.shape)
print("df7",df7.shape)
print("df8",df8.shape)
```

```python
# Los concateno para crear un solo DF y poder trabajar a partir de ese
df_rating = pd.concat([df1,df2,df3,df4,df5,df6,df7,df8])
print(df_rating.shape)
```

```python
# Verifico que se concatenaron correctamente
df_rating
```

```python
# En este punto creo un nuevo dataframe, agrupando el df_rating por "movieId" teniendo en cuenta que el índice del dataframe resultante no se establezca como la columna agrupada
# luego haga un mean o promedio de la columna rating en base a los agrupados y utilize un round para que solo me trajera un numero luego de la coma.
df_grouped = round(df_rating.groupby('movieId', as_index=False)['rating'].mean(),1)
```

```python
# Cambio el nombre de la columna a score
df_grouped = df_grouped.rename(columns={'rating': 'score'})
# Verifico si se realizo correctamente
df_grouped
```

<b>Union de los dataframe<b>

```python
df_plataforma.info()
```

```python
# Uno con un merge dentro del df_plataforma y df_grouped con las columna ID a la izquierda y la columna movieId a la derecha. 
df_plataforma = pd.merge(df_plataforma , df_grouped, left_on='id',right_on='movieId')
```

```python
#Verficio que se aplico.
df_plataforma
```

```python
# Dropeo al columna movieId ya que no la necesito por que tengo la columna ID, lla deje anteriormente para poder verificar que coincidan los id entre ellas.
df_plataforma.drop(columns=['movieId'], inplace= True)
# Verifico si se dropeo correctamente.
df_plataforma.info()
```

```python
# Transformo el dataframe en un CSV para poder utilizarlo
df_plataforma.to_csv('plataformas.csv', index=False, date_format='%Y-%m-%d', encoding="utf-8")
```
